```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
np.random.seed(seed = 16)
```

```python
def brown_motion(x0, n, step):
    res = np.empty(n)
    res[0] = x0
    for i in range(1, n, 1):
        res[i] = res[i-1] + np.random.normal(0, step)
    return res
```

```python
def geom_brown_motion(x0, n, mu, sigma, dt):
    w = brown_motion(x0, n, dt)
    res = np.empty(n)
    res[0] = x0
    for i in range(1, n, 1):
        res[i] = res[0]*np.exp((mu-(sigma**2)*0.5)*i*dt)*np.exp(sigma*w[i])
    return res
```

```python
np.random.seed(seed = 16)
for sg in np.arange(0.8, 2, 0.2):
    plt.plot(geom_brown_motion(1,50,1,sg,0.1))
plt.show()
```
